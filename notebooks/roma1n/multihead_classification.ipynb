```python
remove_extra = lambda x: x.split(' ')[0].split(',')[0]

products = pd.read_csv('../data/lamoda/imagin_public_product.csv', header=None)

dataset = products[[0, 6, 5, 4]]
dataset.columns = ['id', 'link', 'img', 'category']

dataset['img'] = dataset['img'].apply(lambda x: x.split('/')[-1])
dataset['category'] = dataset['category'].apply(remove_extra)

params = pd.read_csv('../data/lamoda/imagin_public_attribute_value.csv', header=None)
params.columns = ['Useless_1', 'param_type', 'id', 'value']

# season == 10
seasons = params[params['param_type'] == 10][['id', 'value']]
seasons.columns = ['id', 'season']
seasons['season'] = seasons['season'].apply(remove_extra)
dataset = pd.merge(left=dataset, right=seasons, how='left', on='id')

# color == 11
colors = params[params['param_type'] == 11][['id', 'value']]
colors.columns = ['id', 'color']
colors['color'] = colors['color'].apply(remove_extra)
dataset = pd.merge(left=dataset, right=colors, how='left', on='id')

# print == 12
prints = params[params['param_type'] == 12][['id', 'value']]
prints.columns = ['id', 'print']
prints['print'] = prints['print'].apply(remove_extra)
dataset = pd.merge(left=dataset, right=prints, how='left', on='id')

# country == 17
countries = params[params['param_type'] == 17][['id', 'value']]
countries.columns = ['id', 'country']
countries['country'] = countries['country']
dataset = pd.merge(left=dataset, right=countries, how='left', on='id')

# brand
brand = products[[0, 1]]
brand.columns = ['id', 'brand_id']
brand_info = pd.read_csv('../data/lamoda/imagin_public_brand.csv', header=None)
brand_info.columns = ['brand_id', 'brand', 'Useless_1']
brand = pd.merge(brand, brand_info, on='brand_id')
brand = brand[['id', 'brand']]
dataset = pd.merge(left=dataset, right=brand, how='left', on='id')
dataset = dataset.dropna()

dataset.head(10)
```
